```python
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image

# Load your image (replace with your image path)
img = Image.open('train/Plantae/f7a74f7568808bc2748f3b4e4623cc4e.jpg')  # or .png, .jpeg, etc.

width, height = img.size

print(f"Original size: {width} x {height}")

# Define the transform
random_crop = transforms.RandomResizedCrop(size=700, scale=(0.5, 1.5))


random_blur = transforms.GaussianBlur(kernel_size=3)

# Apply the transform
transformed_img = random_crop(img)

width, height = transformed_img.size

print(f"Transformed size: {width} x {height}")

# Show original and transformed image side by side
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(img)
axs[0].set_title("Original")
axs[0].axis('off')

axs[1].imshow(transformed_img)
axs[1].set_title("RandomResizedCrop")
axs[1].axis('off')

plt.show()
```
